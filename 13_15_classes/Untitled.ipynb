```python
class Roll:
    """
    Represents a roll

    name: the name of the roll
    wins_against: a list of roll names that this roll wins against
    looses_against: a list of roll names that this roll looses against
    """

    def __init__(self, name, wins_against, looses_against):
        self.name = name
        self.wins_against = wins_against
        self.looses_against = looses_against


class Player:
    """
    Represents a player

    name: the name of the player
    wins: the total number of wins this player has achieved.
    """

    def __init__(self, name):
        self.name = name
        self.wins = 0

    def add_win(self):
        self.wins += 1
```
